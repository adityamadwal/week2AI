```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
```

```python
data=pd.read_csv('ToyotaCorolla.csv')
print(data)
data.head(2)
```

```python
data.dropna(subset=['Model'],inplace=True)
data.isnull().sum()
```

```python
#Visualization for the data
sns.countplot(x=data['Fuel_Type'])
plt.title("Distribution of Passengers by Sex")
plt.show()
```

```python
plt.figure(figsize=(6, 4))
sns.countplot(x='Model', hue='Met_Color', data=data)
plt.title("Model count by Met_Color")
plt.show()
```

```python
x = data["Mfg_Year"]  # X-axis (Years)
y = data["Price"]  # Y-axis (Prices)
z = 0  # All bars at the same depth
colors = plt.cm.inferno(np.linspace(0, 1, len(data)))
# Plotting
fig = plt.figure(figsize=(10, 6))
ax = fig.add_subplot(111, projection="3d")
ax.bar(x, y, zs=z, zdir="z", alpha=0.8, color=colors)
ax.set_xlabel("Mfg_Year")
ax.set_ylabel("Price")
ax.set_zlabel("")  # No Z-axis depth
ax.set_title("3D Bar Plot of Mfg_Year vs Price")
plt.show()
```

```python
n, bins, patches = plt.hist(data["KM"], bins=30, edgecolor='black', alpha=0.7)
colors = plt.cm.viridis(np.linspace(0, 1, len(patches)))
for patch, color in zip(patches, colors):
    patch.set_facecolor(color)
plt.title("KM distributions of Models")
plt.xlabel("KM")
plt.show()
```

```python
sns.violinplot(x=data['Mfg_Month'], y=data['Price'], palette='Spectral')
plt.title("Price Distribution by Mfg_Month")
plt.xlabel("Mfg_Month(0 = No, 1 = Yes)")
plt.ylabel("Price")
plt.show()
```

```python
columns_to_plot = ['Gears', 'Airco', 'Doors', 'Cylinders', 'Boardcomputer']
sns.set_style("whitegrid")
print("Distribution of Features")
fig, axes = plt.subplots(1, len(columns_to_plot), figsize=(18, 5))
for i, column in enumerate(columns_to_plot):
    value_counts = data[column].value_counts()
    axes[i].pie(value_counts, labels=value_counts.index, autopct='%1.1f%%',
                colors=sns.color_palette("pastel", len(value_counts)))
    axes[i].set_title(f"Distribution of {column}")
plt.tight_layout()
plt.show()
```

```python
selected_columns = ['Price', 'HP', 'cc', 'ABS', 'Power_Steering',
                    'Backseat_Divider', 'Metallic_Rim', 'Radio_cassette', 'Sport_Model']
numeric_data = data[selected_columns]
plt.figure(figsize=(10, 6))
sns.heatmap(numeric_data.corr(), annot=True, cmap='RdBu', linewidths=0.5, fmt=".2f")
plt.title("Feature Correlation Heatmap")
plt.show()
```
